```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Problem 2: Investment Problem
plt.figure(figsize=(10, 8))
x = np.linspace(0, 12e6, 1000)

# Plot constraints
plt.plot(x, 12e6 - x, label='Total Budget')
plt.axhline(y=5e6, color='r', label='Max Investment B')
plt.axvline(x=1e7, color='g', label='Max Investment A')
plt.plot(x, 2*x, label='B  A/2')
plt.plot(x, 0.5*x, label='A  B/2')

plt.grid(True, alpha=0.3)
plt.xlabel('Investment in A')
plt.ylabel('Investment in B')
plt.title('Investment Problem')
plt.legend()
plt.show()
```

```python
# Problem 3: Tortilla Problem
plt.figure(figsize=(10, 8))
x = np.linspace(0, 5000, 1000)

# Plot constraints
plt.axvline(x=3200, color= 'y' label='Flour Constraint')
plt.axhline(y=4800, color='r', label='Corn Constraint')
plt.plot(x, 7200 - 2*x, label='Labor Constraint')

plt.grid(True, alpha=0.3)
plt.xlabel('Corn Tortillas')
plt.ylabel('Flour Tortillas')
plt.title('Tortilla Production Problem')
plt.legend()
plt.show()
```

```python
# Problem 4: Robot Problem
plt.figure(figsize=(10, 8))
x = np.linspace(0, 60, 1000)

# Plot constraints
plt.axhline(y=30, color='r', label='Max Deluxe')
plt.axvline(x=50, color='g', label='Max Basic')
plt.plot(x, 70 - x, label='Assembly Constraint')
plt.plot(x, (60 - 0.5*x)/1.6, label='Programming Constraint')

plt.grid(True, alpha=0.3)
plt.xlabel('Basic Robots')
plt.ylabel('Deluxe Robots')
plt.title('Robot Production Problem')
plt.legend()
plt.show()
```

```python
# Problem 5: Original Case
plt.figure(figsize=(10, 8))
x = np.linspace(0, 2000, 1000)

# Plot constraints
plt.plot(x, (10000 - x)/1.5, label='Resource 1')
plt.plot(x, (4600 - 3*x)/2, label='Resource 2')

plt.grid(True, alpha=0.3)
plt.xlabel('Product 1')
plt.ylabel('Product 2')
plt.title('Product Problem - Original')
plt.legend()
plt.show()
```

```python
# Problem 5: Modified Case
plt.figure(figsize=(10, 8))
x = np.linspace(0, 2000, 1000)

# Plot constraints
plt.plot(x, (12000 - x)/1.5, label='Resource 1')
plt.plot(x, (4200 - 3*x)/2, label='Resource 2')

plt.grid(True, alpha=0.3)
plt.xlabel('Product 1')
plt.ylabel('Product 2')
plt.title('Product Problem - Modified')
plt.legend()
plt.show()
```
